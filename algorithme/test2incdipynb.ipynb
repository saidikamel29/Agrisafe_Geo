```python
#import onnée FIRMS
import pandas as pd

API_KEY = "fda7b9bf8a7ff50df7cdd64fb8b12a16"

url = f"https://firms.modaps.eosdis.nasa.gov/api/area/csv/{API_KEY}/VIIRS_NOAA20_NRT/-8,18,12,38/3"

df = pd.read_csv(url)

print(df.head())
print("Number of fire points:", len(df))
```

```python
#transfert lant et lang vers point geomitrie 
import geopandas as gpd
from shapely.geometry import Point

# تحويل dataframe إلى GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=[Point(xy) for xy in zip(df.longitude, df.latitude)],
    crs="EPSG:4326"
)

print(gdf.head())
```

```python
polygon_coords = [
    (3.0, 35.7),
    (3.1, 35.7),
    (3.1, 35.6),
    (3.0, 35.8)
]
```

```python
from shapely.geometry import Polygon

field_polygon = Polygon(polygon_coords)
```

```python
#  BUFFER 
buffer_polygon = field_polygon.buffer(0.005)  # ~500m
```

```python
fires_inside = gdf[gdf.geometry.within(buffer_polygon)]

print("Fires detected:", len(fires_inside))
```

```python
result = {
    "fire_detected": len(fires_inside) > 0,
    "fire_count": int(len(fires_inside))
}

print(result)
```
